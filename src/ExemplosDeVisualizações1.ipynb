## Scatterplot

Os dados são do dataset de House Prices do Kaggle (https://www.kaggle.com/c/house-prices-advanced-regression-techniques/data).

Dados: subset, apenas para observar a relação entre preço e a área.



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../raw/housePrices.csv")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data['GrLivArea'], data['SalePrice']/1000, alpha=0.5)

plt.xlabel('Ground Living Area')
plt.ylabel('Sale Price (k$)')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, alpha=0.8, color="#69b3a2")
sns.regplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, scatter=False, line_kws={'alpha': 0.8, 'lw': 0.5, 'color': 'skyblue'});
plt.xlabel('Ground Living Area')
plt.ylabel('Sale Price (k$)')
plt.show()
```

Overplotting

```python
f, axes = plt.subplots(2, 2, figsize=(20, 20))
sns.scatterplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, alpha=0.8, color="#69b3a2", ax=axes[0, 0])
sns.kdeplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, cmap='Blues', ax=axes[0, 1])
sns.kdeplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, cmap='Blues', fill=True, ax=axes[1, 0])
sns.kdeplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, cmap='viridis', fill=True, ax=axes[1, 1])
plt.show()
```

```python
sns.jointplot(data=data, x='GrLivArea', y=data['SalePrice']/1000, kind='hist', cmap='viridis', marginal_kws={'color': 'skyblue'})
plt.show()
```
